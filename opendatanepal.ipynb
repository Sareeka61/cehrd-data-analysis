```python
import pandas as pd
```

```python
df =pd.read_csv('housingdata.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.sort_values("District", ascending = False)
```

```python
df = df.sort_values("Total population", ascending = False)
```

```python
grouped_df = df.groupby(by="District")["Total population"].sum().reset_index()
grouped_df.sort_values(by="Total population", inplace=True)
```

```python
grouped_df.head(10)
```

```python
grouped_df.sort_values(by="Total population", ascending=False)
```

```python
for district, district_df in grouped_df.groupby(by="District"):
    print(district)
    print(district_df.head(10))
```

```python
least_populated_district = grouped_df.head(5)
```

```python
least_populated_district.info()
```

```python
least_populated_district.reset_index(inplace=True, drop=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(least_populated_district["District"], least_populated_district["Total population"], marker='o')
plt.xlabel("District")
plt.ylabel("Total population")
plt.title("Least populated districts")
plt.show()
```
