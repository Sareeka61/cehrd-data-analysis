```python
import pandas as pd
```

```python
# Sample data
data ={
    'A' : [1,2,2,3,4,5,5,5,6],
    'B' : [7,8,9,10,11,12,13,14,15],
    'C' : [16,17,18,19,20,21,22,23,24]
    }
```

```python
# Creating a dataframe
df = pd.DataFrame(data)
```

```python
df.head(9)
```

```python
df.tail(9)
```

```python
mode_a = df['A'].mode()
mode_a[0]
mode_b = df['B'].mode()[0]
mode_c = df['C'].mode()[0]

print(f"Mode of A: {mode_a[0]}")
print(f"Mode of B: {mode_b}")
print(f"Mode of C: {mode_c}")
```

```python
mean_a = df['A'].mean()
mean_b = df['B'].mean()
mean_c = df['C'].mean()

print(f"Mean of A: {mean_a}")
print(f"Mean of B: {mean_b}")
print(f"Mean of C: {mean_c}")
```

```python
median_a = df['A'].median()
median_b = df['B'].median()
median_c = df['C'].median()

print(f"Median of A: {median_a}")
print(f"Median of B: {median_b}")
print(f"Median of C: {median_c}")
```

```python
#Calculating standard deviation

std_A = df['A'].std()
std_B = df['B'].std()
std_C = df['C'].std()

print(f"Standard Deviation of A: {std_A}")
print(f"Standard Deviation of B: {std_B}")
print(f"Standard Deviation of C: {std_C}")
```
